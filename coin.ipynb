```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Flip a (fair) coin 'n' times.
rng = np.random.default_rng(123123)

def flip_coin(n):
    rints = rng.integers(low=0, high=2, size=n) # Note: high is not included.
    return rints

flip_coin(10)
```

```python
# Demonstrates the LLN for the simple case of a coin flip.  

plt.plot([flip_coin(n).mean() for n in range(1, 10000)])
```

```python
# Demonstrating the CLT via simulation for the simple case of a coin flip.
# Unnormalized form.

n = 100

def unique_values_probabilities(x):
    values, counts = np.unique(x, return_counts=True)
    probs = counts / len(x)
    return values, probs

# sigma = 1/2*((0 - 0.5)**2 + (1 - 0.5)**2) 
mu, sigma = 0.5, 0.25

number_samples = 10000
coin_flips = [flip_coin(n) for i in range(number_samples)]
sums = [sum(d) for d in coin_flips]
#normalized = [(s -  n*mu) / (sigma * math.sqrt(n)) for s in sums]

values, probs = unique_values_probabilities(sums)

plt.plot(values, probs)
```

```python
#normalized = [(s -  n*mu) / (sigma * math.sqrt(n)) for s in sums]
normalized = [( (s / n) -  mu) * math.sqrt(n) for s in sums]
values, probs = unique_values_probabilities(normalized)

plt.plot(values, probs)
```

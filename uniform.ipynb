```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Continuous uniform distribution example.

def sample_continuous_uniform(a, b, n, number_samples):
# Endpoints are 'a' and 'b', 'n' is the number of random variables.
# 'number_samples' is the number of samples taken to approximate the distribution.
    return [sum(np.random.uniform(a, b, n)) for i in range(number_samples)]

def sample_continuous_uniform_normalized(a, b, n, number_samples):
    mu, sigma = 0.5*(b - a), (b - a) / math.sqrt(12)
    sum_samples = sample_continuous_uniform(a, b, n, number_samples)
    return [(s -  n*mu) / (sigma * math.sqrt(n)) for s in sum_samples]
```

```python
# (Not yet normalized) graphs for continuous uniform distribution.

a, b = 0, 10
n_values = [1, 2, 5, 10, 20]
number_samples = 1000
bins = 10

fig, axes = plt.subplots(len(n_values))
fig.subplots_adjust(top=2)
for i,n in zip(range(len(n_values)), n_values):
    axes[i].hist(sample_continuous_uniform(a, b, n, number_samples),
                 bins=bins, color="red", alpha=0.5, label=f"n is {n}", density=True)
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Percent per unit")
    axes[i].legend()
```

```python
# Normalized graphs for continuous uniform distribution.

a, b = 0, 10
num_figs = 5
n = 20
number_samples = 1000
bins = 10

fig, axes = plt.subplots(2, num_figs, sharex=True, sharey=True)
fig.subplots_adjust(right=2)
for i in range(num_figs):
    axes[0, i].hist(sample_continuous_uniform_normalized(a, b, n*(i+1), number_samples),
                    bins=bins, color="red", alpha=0.5, label=f"n = {n*(i+1)}", density=True)
    axes[1, i].hist(sample_continuous_uniform_normalized(a, b, n, number_samples*(i+1)),
                    bins=bins, color="red", alpha=0.5, label=f"{number_samples*(i+1)} samples", density=True)
    axes[0, i].legend(loc="upper right")
    axes[1, i].legend(loc="upper right")    
    axes[0, i].set_xlabel("Value")
    axes[1, i].set_xlabel("Value")
    if i == 0:
        axes[0, i].set_ylabel("Percent per unit")
        axes[1, i].set_ylabel("Percent per unit")
```

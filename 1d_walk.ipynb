```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nsteps = 1000

rng = np.random.default_rng(seed=12345)
draws = rng.integers(0, 2, size=nsteps)
steps = np.where(draws == 0, -1, 1) # Changes all of the 0's into -1's.
walk = steps.cumsum()
plt.plot(walk[:100])
```

```python
print("min of walk: " + str(walk.min()))
print("max of walk: " + str(walk.max()))
print("final value: " + str(walk[-1]))
```

```python
# Simulation of CLT for 1-dimensional random walks.

nsteps = 1000
nwalks = 50000

draws = rng.integers(0, 2, size=(nwalks, nsteps))
steps = np.where(draws == 0, -1, 1)
walks = steps.cumsum(axis=1)

ends_of_walks = walks[:,nsteps-1]
```

```python
# Demonstrating the CLT via simulation for the case of a 1-dimensional random walk.
# Unnormalized form.

def unique_values_probabilities(x):
    values, counts = np.unique(x, return_counts=True)
    probs = counts / len(x)
    return values, probs

# sigma = 1/2*((1 - 0)**2 + (-1 - 0)**2) 
mu, sigma = 0.5, 1.0

values, probs = unique_values_probabilities(ends_of_walks)

plt.plot(values, probs)
```
